# Creating a Data Frame
- Data enclosed in a Dictionary


```python
# create DataFrame in pandas

# We can write everything in single or double quotes 
mydata = pd.DataFrame({
    'Name': ['Naruto', 'Ichigo', 'Luffy', 'Goku'],
    'Anime': ["Naruto", "Bleach", "One Piece", "Dragon Ball"],
    "Studio": ['Pierrot', 'Pierrot', 'Toei Animation', 'Toei Animation'], 
    "Episodes": [220, 366, 1000, 153],
    "Rating": [8.3, 7.9, 8.8, 8.5]
})

# And in a single line too 

fruits = pd.DataFrame({'Fruits':['Apples','Mangoes','Grapes','Oranges'],"Quantity":[100,360,20,180]})
```
